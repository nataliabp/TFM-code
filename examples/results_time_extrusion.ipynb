```python
import numpy as np
import os
import matplotlib.pyplot as plt
from scipy.stats import sem
```

```python
def read_results_extrusion(m, date, size): 
    # m: String, level of mutant cells
    # date: String, date of the simulation
    root= r'C:\Users\natal\OneDrive\Documentos\MASTER\TFM\Simulations\Time_extrusion'
    folder = 'sim_' + m +'_'+size+'_'+ date
    path_folder = os.path.join(root, folder)

    m0 = []; std0 =[]; m05 = []; std05 =[]; m07 = []; std07 = []

    # NOISE = 0
    m0_path = os.path.join(path_folder, '0mean.txt')
    with open(m0_path,'r') as area_file: 
        lines = area_file.readlines()
        m0 = [float(line.strip()) for line in lines]

    std0_path = os.path.join(path_folder, '0std.txt')
    with open(std0_path,'r') as area_file: 
        lines = area_file.readlines()
        std0 = [float(line.strip()) for line in lines]
    
    # NOISE = 0.25
    m025_path = os.path.join(path_folder, '0.25mean.txt')
    with open(m025_path,'r') as area_file: 
        lines = area_file.readlines()
        m025 = [float(line.strip()) for line in lines]

    std025_path = os.path.join(path_folder, '0.25std.txt')
    with open(std025_path,'r') as area_file: 
        lines = area_file.readlines()
        std025 = [float(line.strip()) for line in lines]

    
    # NOISE = 0.5
    m05_path = os.path.join(path_folder, '0.5mean.txt')
    with open(m05_path,'r') as area_file: 
        lines = area_file.readlines()
        m05 = [float(line.strip()) for line in lines]

    std05_path = os.path.join(path_folder, '0.5std.txt')
    with open(std05_path,'r') as area_file: 
        lines = area_file.readlines()
        std05 = [float(line.strip()) for line in lines]

    # NOISE = 0.7
    m07_path = os.path.join(path_folder, '0.7mean.txt')
    with open(m07_path,'r') as area_file: 
        lines = area_file.readlines()
        m07 = [float(line.strip()) for line in lines]

    std07_path = os.path.join(path_folder, '0.7std.txt')
    with open(std07_path,'r') as area_file: 
        lines = area_file.readlines()
        std07 = [float(line.strip()) for line in lines]

    return m0, std0, m025, std025, m05, std05, m07, std07

```

```python

m0, std0,m025, std025, m05, std05, m07, std07 = read_results_extrusion('0.20', '7sep', '18')
```
