```python
import numpy as np
import os
import matplotlib.pyplot as plt
from scipy.stats import sem
```

```python
def read_results_shape_index(m, noise): 
    # m: String, level of mutant cells
    # date: String, date of the simulation
    root= r'C:\Users\natal\OneDrive\Documentos\MASTER\TFM\Simulations\ShapeIndex'
    folder = noise + '_' + m 
    path_folder = os.path.join(root, folder)

    si =[]; et = []; co = []

    si_path = os.path.join(path_folder, 'shape_index.txt')
    with open(si_path,'r') as area_file: 
        lines = area_file.readlines()
        si = [float(line.strip()) for line in lines]

    et_path = os.path.join(path_folder, 'extrusion_time.txt')
    with open(et_path,'r') as area_file: 
        lines = area_file.readlines()
        et = [float(line.strip()) for line in lines]
    
    co_path = os.path.join(path_folder, 'cells_out.txt')
    with open(co_path,'r') as area_file: 
        lines = area_file.readlines()
        co = [float(line.strip()) for line in lines]

    return si,et, co
```

```python
def read_results_all(m, noise): 
    # m: String, level of mutant cells
    # date: String, date of the simulation
    root= r'C:\Users\natal\OneDrive\Documentos\MASTER\TFM\Simulations\ShapeIndex'
    folder = noise + '_' + m 
    path_folder = os.path.join(root, folder)

    si =[]; et = []; co = []; a = []; e = []; ad = []; ld = []

    si_path = os.path.join(path_folder, 'shape_index.txt')
    with open(si_path,'r') as area_file: 
        lines = area_file.readlines()
        si = [float(line.strip()) for line in lines]

    et_path = os.path.join(path_folder, 'extrusion_time.txt')
    with open(et_path,'r') as area_file: 
        lines = area_file.readlines()
        et = [float(line.strip()) for line in lines]
    
    co_path = os.path.join(path_folder, 'cells_out.txt')
    with open(co_path,'r') as area_file: 
        lines = area_file.readlines()
        co = [float(line.strip()) for line in lines]
    
    a_path = os.path.join(path_folder, 'area.txt')
    with open(a_path,'r') as area_file: 
        lines = area_file.readlines()
        a = [float(line.strip()) for line in lines]

    e_path = os.path.join(path_folder, 'edges.txt')
    with open(e_path,'r') as area_file: 
        lines = area_file.readlines()
        e = [float(line.strip()) for line in lines]

    ad_path = os.path.join(path_folder, 'angle_distortion.txt')
    with open(ad_path,'r') as area_file: 
        lines = area_file.readlines()
        ad = [float(line.strip()) for line in lines]

    ld_path = os.path.join(path_folder, 'length_distortion.txt')
    with open(ld_path,'r') as area_file: 
        lines = area_file.readlines()
        ld = [float(line.strip()) for line in lines]
        
    return si,et, co, a, e, ld, ad 
```
